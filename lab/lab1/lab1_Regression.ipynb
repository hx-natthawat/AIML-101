## 2. Building the Linear Regression Model

Linear regression finds the best-fitting line through our data points using the equation:
y = mx + b

Where:
- y is the dependent variable (Fuel Cost)
- x is the independent variable (Number of Passengers)
- m is the slope (coefficient)
- b is the y-intercept

The model will find the optimal values for m and b that minimize the sum of squared residuals.
